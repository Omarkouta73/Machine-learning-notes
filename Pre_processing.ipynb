# Data Preprocessing Tutorial

### First let's import our dataset

```python
df = pd.read_csv('unprocessed_data.csv')
df.head()
```

```python
df.describe()
```

```python

y= df['Purchased']
X=df.drop(['Purchased'],1)
X
```

## Start with First Column its Categorical so Lets Turn it into  Dummies

```python

X=pd.get_dummies(X)
X
```

## Handling the missing data
 we have two options :
 1. remove the row
 It can be done easily through excel or pandas or numpy by slicing
 2. replace it by the mean
 It also can be done with many ways but lets use Imputer

```python
missing=df.isnull().sum()
print(missing)
```

```python
missing_prec=100*df.isnull().sum()/len(df)
missing_prec
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values= np.nan, strategy='mean')

X = imputer.fit_transform(X.values)
X
```

```python

X=X.astype(int)
X
```
