# Clustering - How k-means work ?

*Lets consider we have points P1(1,3) , P2(2,2) , P3(5,8) , P4(8,5) , P5(3,9) , P6(10,7) , P7(3,3) , P8(9,4) , P9(3,7).*

## Step 1: Calculate the number of clusters 

*The Elbow Method*

**Say we have 3 clusters. K=3**

## Step 2: Select random K data points as cluster center

*We assume that our Initial cluster centers are P7(3,3), P9(3,7), P8(9,4) as C1, C2, C3*

## Step 3: Calculate the distance between each data point and the centers

***Choose the cluster center for each data point by calculating the euclidean distance***

### Iteration 1

*Choose the cluster center for P1 (1,3)*

* ***C1P1 = (3,3) and (1,3) = sqrt([1-3]^2+[3-3]^2) = 2***
* ***C2P1 = (3,7) and (1,3) = sqrt([1-3]^2+[3-7]^2) = 4.5***
* ***C3P1 = (9,4) and (1,3) = sqrt([1-9]^2+[3-4]^2) = 8.1***

***This means that P1 (1,3) is related to the first center (3,3)***

*Choose the cluster center for P2 (2,2)*

* ***C1P2 = (3,3) and (2,2) = sqrt([2-3]^2+[2-3]^2) = 1.4***
* ***C2P2 = (3,7) and (2,2) = sqrt([2-3]^2+[2-7]^2) = 5.1***
* ***C3P2 = (9,4) and (2,2) = sqrt([2-9]^2+[2-4]^2) = 7.3***

***This means that P2 (2,2) is related to the first center (3,3)***

*Choose the cluster center for P3 (5,8)*

* ***C1P3 = (3,3) and (5,8) = sqrt([5-3]^2+[8-3]^2) = 5.3***
* ***C2P3 = (3,7) and (5,8) = sqrt([5-3]^2+[8-7]^2) = 2.2***
* ***C3P3 = (9,4) and (5,8) = sqrt([5-9]^2+[8-4]^2) = 5.7***

***This means that P3 (5,8) is related to the second center (3,7)***

### And Continue the same way...

* ***Cluster 1 => P1(1,3) , P2(2,2) , P7(3,3)***
* ***Cluster 2 => P3(5,8) , P5(3,9) , P9(3,7)***
* ***Cluster 3 => P4(8,5) , P6(10,7) , P8(9,4)***

### Iteration 2

*Re-compute the new clusters and the new cluster center is computed by taking the mean of all the points contained in that particular cluster.*

* ***New center of Cluster 1 => (1+2+3)/3=2 , (3+2+3)/3=2.7 => (2,2.7)***
* ***New center of Cluster 2 => (5+3+3)/3=3.7 , (8+9+7)/3=8 => (3.7,8)***
* ***New center of Cluster 3 => (8+10+9)/3=9 , (5+7+4)/3=5.3 => (9,5.3)***

**Re-calculate the distance between data points and new cluster centers**

### Results

**The table shows each point and what centroid does the point follow**

* ***Cluster 1 => P1(1,3) , P2(2,2) , P7(3,3)***
* ***Cluster 2 => P3(5,8) , P5(3,9) , P9(3,7)***
* ***Cluster 3 => P4(8,5) , P6(10,7) , P8(9,4)***

*K-Means clustering stops iteration because of the same cluster repeating so no need to continue iteration and display the last iteration as the best cluster groups for this dataset.*
